# **Working with Images**
* Ploting Images using Matplotlib.
* Matplotlib library supports the basic operations on images like loading, rescaling, and displaying the images.



```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
```

```python
img = image.load_img("laptop/1.jpg")
plt.imshow(img)
```

```python
import os

dir_path = "laptop"
for i in os.listdir(dir_path):
  if '.jpg' in i:
    img = image.load_img(dir_path+"//"+i,target_size=(250,250))
    plt.imshow(img)
    plt.show()
```
