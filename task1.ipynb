```python
import matplotlib.pyplot as plt
import requests

uk = requests.get("http://universities.hipolabs.com/search?country=United+Kingdom")
ukdata = uk.json()

us = requests.get("http://universities.hipolabs.com/search?country=United+States")
usdata = us.json()

ind = requests.get("http://universities.hipolabs.com/search?country=india")
inddata = ind.json()

pak = requests.get("http://universities.hipolabs.com/search?country=pakistan")
pakdata = pak.json()

countries = ['United Kingdom', 'United States', 'India', 'Pakistan']
totaluniversities = [len(ukdata), len(usdata), len(inddata), len(pakdata)]

fig = plt.figure(figsize=(15,10))
plt.bar(countries, totaluniversities)
plt.xlabel("COUNTRIES")
plt.ylabel("TOTAL UNIVERSITIES")
plt.show()

# ----------Search Mechanism---------------- # Close the Bar Graph Window to run and check search mechanism output

inuputcountry = input('Enter Country Name: ')
inuputuniversity = input('Enter University Name: ')
n = 0
data = []
if inuputcountry.lower() == 'india':
    n = len(inddata)
    data = inddata
elif inuputcountry.lower() == 'united kingdom':
    n = len(ukdata)
    data = ukdata
elif inuputcountry.lower() == 'united states':
    n = len(usdata)
    data = usdata
elif inuputcountry.lower() == 'pakistan':
    n = len(pakdata)
    data = pakdata
else:
    print('Please enter a valid country name')
    exit()

for i in range(0,n):
    if inuputcountry.lower() == data[i]['country'].lower() and inuputuniversity.lower() == data[i]['name'].lower():
        print('University Found', ' Visit - ', data[i]['web_pages'])
        exit()
else:
    print('University Not Found')
```
